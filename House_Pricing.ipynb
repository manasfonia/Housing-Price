## Seperating Numerical and Object type Data

```python
Data_num = Data.select_dtypes(include=["float64", "int64"]).copy()
Data_obj = Data.select_dtypes(include ="object").copy()
```

```python
del Data_num['Id']
```

```python
n = pd.DataFrame(Data_num.isnull().sum())
n.columns=["c"]
for i in range(len(n["c"])):
    if n["c"][i]>0:
        print(n.index[i],' ',n["c"][i])
```

```python
o = pd.DataFrame(Data_obj.isnull().sum())
o.columns=["c"]
for i in range(len(o["c"])):
    if o["c"][i]>0:
        print(o.index[i],' ',o["c"][i])
```

```python
imp = IterativeImputer()
Data_num = pd.DataFrame(imp.fit_transform(Data_num), columns = Data_num.columns)
```

```python
Data_obj.replace(np.nan,'Unknown', inplace=True)
```

## Feature Scaling

```python
scaler = MinMaxScaler()
Data_num = pd.DataFrame(scaler.fit_transform(Data_num), columns = Data_num.columns)
```

```python
for i in Data_obj.columns:
    Data_obj[i]=Data_obj[i].astype('category')
```

```python
for i in Data_obj.columns:
    print(Data_obj[i].value_counts())
```

```python
Data_obj_onehot = Data_obj.copy()
for i in Data_obj.columns:
    Data_obj_onehot = pd.get_dummies(Data_obj_onehot, columns=[i], prefix_sep="_|_")
```

```python
Data_obj_onehot
```

```python
Data_final = pd.concat([Data_num,Data_obj_onehot], axis=1, sort=False)
```

```python
Data_final.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(Data_final, y, test_size =0.1, random_state=100)
```

```python
X_train.shape
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import max_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
def generate_model_report(y_actual, y_predicted):
    print("r2 score = ", r2_score(y_actual, y_predicted))
    print("Max Error = ", max_error(y_actual, y_predicted))
    print("Mean Absolute Erroe = ", mean_absolute_error(y_actual, y_predicted))
    print("Mean Squared Error = ", mean_squared_error(y_actual, y_predicted))
    print("Mean squared logarithmic error = ", mean_squared_log_error(y_actual, y_predicted))
    pass
```

```python
ols = RandomForestRegressor()
model = ols.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
generate_model_report(Y_test, Y_pred)
```

```python
ols = DecisionTreeRegressor()
model = ols.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
generate_model_report(Y_test, Y_pred)
```

```python
las = Lasso()
model = las.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
generate_model_report(Y_test, Y_pred)
```

```python
rid = Ridge()
model = rid.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
generate_model_report(Y_test, Y_pred)
```
